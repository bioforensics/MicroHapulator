```python
profiles = list()
for i in range(10):
    pop = random.choice(pops)
    seed = random.randint(0, 2**32-1)
    p = microhapulator.sim.sim([pop], panel=markers, seed=seed)
    profiles.append(p)
```

```python
profiles[0].data["metadata"]
```

```python
tables = list()
for profile in profiles:
    rmps = dict(
        Population=list(),
        RandMatchProb=list(),
    )
    for pop in pops:
        p = profile.rand_match_prob(pop)
        rmps["Population"].append(pop)
        rmps["RandMatchProb"].append(p)
    d = pandas.DataFrame(rmps)
    d["Actual"] = profile.data["metadata"]["MaternalHaploPop"]
    d = d.sort_values("RandMatchProb", ascending=False)[["Actual", "Population", "RandMatchProb"]]
    tables.append(d)
```

```python
for t in tables:
    display(t)
```

```python
for n, t in enumerate(tables):
    filename = f"table{n}.csv"
    t.to_csv(filename, index=False)
```
